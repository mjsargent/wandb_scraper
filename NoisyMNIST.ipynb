```python
import matplotlib.pyplot as plt
from scraper import Line

plt.rc('font', family='serif')
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
```

```python
AMA_zeros = Line(
    run_names=[
        "sweet-plasma-12",
        "polar-monkey-11",
        "pretty-wildflower-10",
    ],
    x_quantity="_step",
    y_quantity="zeros loss",
    color="#95275F",
    project="Noisy MNIST"
)

AMA_ones = Line(
    run_names=[
        "sweet-plasma-12",
        "polar-monkey-11",
        "pretty-wildflower-10",
    ],
    x_quantity="_step",
    y_quantity="ones loss",
    color="#229487",
    project="Noisy MNIST"
)

MSE_zeros = Line(
    run_names=[
        "hardy-serenity-15",
        "still-rain-14",
        "toasty-elevator-13",
    ],
    x_quantity="_step",
    y_quantity="zeros loss",
    color="#EDB732",
    project="Noisy MNIST"
)

MSE_ones = Line(
    run_names=[
        "hardy-serenity-15",
        "still-rain-14",
        "toasty-elevator-13",
    ],
    x_quantity="_step",
    y_quantity="ones loss",
    color="#5387DD",
    project="Noisy MNIST"
)
```

```python
AMA_zeros.plot_line(label="AMA Deterministic")
AMA_ones.plot_line(label="AMA Stochastic")
MSE_zeros.plot_line(label="MSE Deterministic")
MSE_ones.plot_line(label="MSE Stochastic")

# make graph pretty
plt.xlabel("Training Steps", fontsize="x-large")
plt.ylabel("Intrinsic Rewards", fontsize="x-large")
plt.xlim(-5,1000)
plt.ylim(-0.03,0.11)
plt.title("Noisy MNIST Intrinsic Rewards", fontsize="xx-large")
plt.tight_layout()
plt.legend()
plt.savefig("NoisyMNISTIntRewards.png", dpi=1200)
plt.show()
```
