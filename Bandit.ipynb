```python
import matplotlib.pyplot as plt
from scraper import Line
import wandb
import math
from scipy import stats
import numpy as np

plt.rc('font', family='serif')
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
```

```python
noisy_epistemic_ACh = Line(
    run_names=[
        "fresh-glitter-56",
        "swift-lake-57",
        "crimson-salad-58",
    ],
    x_quantity="_step",
    y_quantity="noisy_intrinsic_reward",
    color="#FF0DAE",
    project="bandit",
)

stable_epistemic_ACh = Line(
    run_names=[
        "fresh-glitter-56",
        "swift-lake-57",
        "crimson-salad-58",
    ],
    x_quantity="_step",
    y_quantity="stable_intrinsic_reward",
    color="#6CA800",
    project="bandit",
)
```

```python
noisy_epistemic_ACh.plot_line(label="Stochastic Zone")
stable_epistemic_ACh.plot_line(label="Deterministic Zone")

plt.xlim(0, 6e4)
plt.ylim(0, 0.03)
# make graph pretty
plt.xlabel("Training Steps", fontsize="x-large")
plt.ylabel("ACh Level", fontsize="x-large")
plt.title("Epistemic Model of ACh", fontsize="xx-large")
plt.tight_layout()
plt.savefig("EpistemicACh.png", dpi=1200)
plt.legend()
plt.show()
```

```python
noisy_aleatoric_ACh = Line(
    run_names=[
        "crisp-field-55",
        "curious-voice-54",
        "glowing-wind-53",
    ],
    x_quantity="_step",
    y_quantity="noisy uncertainty",
    color="#229487",
    project="bandit",
)
stable_aleatoric_ACh = Line(
    run_names=[
        "crisp-field-55",
        "curious-voice-54",
        "glowing-wind-53",
    ],
    x_quantity="_step",
    y_quantity="stable uncertainty",
    color="#3C23E8",
    project="bandit",
)
```
