```python
import matplotlib.pyplot as plt
from scraper import Line

plt.rc('font', family='serif')
plt.rc('xtick', labelsize='large')
plt.rc('ytick', labelsize='large')
```

```python
eta_one_lambda_small = Line(
    run_names=[
        "vocal-eon-1296",
        "soft-cloud-1316",
        "polished-haze-1325",
    ],
    x_quantity="tcount",
    y_quantity="eprew",
    color="#95275F",
    project="burda_pathak_fork"
)

eta_one_lambda_medium = Line(
    run_names=[
        "azure-voice-1299",
        "dauntless-breeze-1317",
        "super-fog-1326",
    ],
    x_quantity="tcount",
    y_quantity="eprew",
    color="#95275F",
    project="burda_pathak_fork"
)

eta_one_lambda_large = Line(
    run_names=[
        "jolly-monkey-1301",
        "proud-valley-1318",
        "cosmic-fog-1327",
    ],
    x_quantity="tcount",
    y_quantity="eprew",
    color="#95275F",
    project="burda_pathak_fork"
)
```

```python
eta_one_lambda_small.plot_line()
eta_one_lambda_medium.plot_line()
eta_one_lambda_large.plot_line()

# make graph pretty
plt.xlabel("Training Steps", fontsize="x-large")
plt.ylabel("Extrinsic Reward", fontsize="x-large")
plt.title("$\eta=1$, Variable $\lambda$", fontsize="xx-large")
plt.tight_layout()
plt.xlim(0,1.5e6)
plt.ylim(0,1100)
plt.savefig("variable_lambda.png", dpi=1200)
plt.show()
```
