```python
import matplotlib.pyplot as plt
from scraper import Line

plt.rc('font', family='serif')
plt.rc('ytick', labelsize='large')
```

```python
idf_bank_heist = Line(
    run_names=[
        "good-thunder-1050",
        "spring-energy-1072",
        "polished-monkey-1078",
    ],
    x_quantity="_step",
    y_quantity="average states visited",
    color="#EDB732",
    project="burda_pathak_fork"
)
idf_AMA_bank_heist = Line(
    run_names=[
        "hearty-sun-1077",
        "skilled-bush-1082",
        "dandy-pyramid-1644",
    ],
    x_quantity="_step",
    y_quantity="average states visited",
    color="#FF66CC",
    project="burda_pathak_fork"
)
```

```python
idf_bank_heist.plot_line(label="IDF")
idf_AMA_bank_heist.plot_line(label="IDF AMA")


# make graph pretty
plt.xlabel("Training Steps", fontsize="x-large")
plt.ylabel("Average States Visited", fontsize="x-large")
plt.xlim(0,3000)
# plt.ylim(0,420)
plt.title("Bank Heist Maze Coverage", fontsize="xx-large")
plt.tight_layout()
plt.savefig("BankHeist.png", dpi=1200)
plt.legend()
plt.show()
```
